```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mn = 1
mx = 10
size = 5
x = np.random.choice(range(mn, mx+1), size, replace=False)
y = np.random.choice(range(mn, mx+1), size, replace=False)
```

```python
x, y
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel('x-axis', fontdict={'c':'blue', 'size':15})
plt.ylabel('y-axis', fontdict={'c':'blue', 'size':15})
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel('x axis', fontdict={'c':'blue', 'size':15})
plt.ylabel('y axis', fontdict={'c':'blue', 'size':15})
plt.xlim(0,max(x)+1)
plt.ylim(0,max(y)+1)
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel('x axis', fontdict={'c':'blue', 'size':15})
plt.ylabel('y axis', fontdict={'c':'blue', 'size':15})
plt.xlim(0,max(x)+1)
plt.ylim(0,max(y)+1)
plt.xticks(range(0,max(x)+1,1))
plt.yticks(range(0,max(y)+1,1))
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel('x axis', fontdict={'c':'blue', 'size':15})
plt.ylabel('y axis', fontdict={'c':'blue', 'size':15})
plt.xlim(0,max(x)+1)
plt.ylim(0,max(y)+1)
plt.xticks(range(0,max(x)+1,1))
plt.yticks(range(0,max(y)+1,1))
plt.hlines(y, 0, x, linestyles='dotted')
plt.vlines(x, 0, y, linestyles='dotted')
plt.show()
```

```python
plt.figure(figsize=(5,10))
plt.scatter(x, y)
plt.xlabel('x axis', fontdict={'c':'blue', 'size':15})
plt.ylabel('y axis', fontdict={'c':'blue', 'size':15})
plt.xlim(0,max(x)+1)
plt.ylim(0,max(y)+1)
plt.xticks(range(0,max(x)+1,1))
plt.yticks(range(0,max(y)+1,1))
plt.hlines(y, 0, x, linestyles='dotted')
plt.vlines(x, 0, y, linestyles='dotted')
plt.show()
```
