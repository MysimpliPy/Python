```python
import pandas as pd
```

```python
term  =  ['first', 'second']
names = ['Ravi', 'Naveen', 'Hema', 'Lata']
subject = ['Math', 'Chemistry', 'Physics', 'Biology', 'English']
marks  = np.random.randint(0, 101, 40)
```

```python
cols = [ [t,n,s] for t in term for n in names for s in subject]   # cartesian
```

```python
df = pd.DataFrame(cols, columns=['term', 'name', 'subject'])
```

```python
df['marks'] = marks
```

```python
df.head(2)
```

```python
# list out name-wise total marks obtained in each term

# select term, name, sum(marks) from df   # df = table
# group by term, name

df.groupby(['term', 'name']).sum()   # min, max, mean
```

```python
# total marks obtained by each student

# select name, sum(marks) from df 
# group by name


df.groupby(['name']).sum()  
```

```python
for i in np.unique(df.name) :
    print(df.loc[(df.name == i), :].groupby('name').sum())
    print()
```

```python
# list out name-wise total marks obtained in each term

# select term, name, sum(marks) from df   # df = table
# group by term, name

df[['term', 'name', 'marks']].groupby(['term', 'name']).sum()   # min, max, mean
```

```python
# average marks of each term for all students

# select term, name, avg(marks) from df 
# group by term, name

df.groupby(['term', 'name']).mean()
```

```python
# find out average marks of each term for all students 
# and get the output in ascending order of marks

# select term, name, avg(marks) from df 
# group by term, name
# order by marks

df.groupby(['term', 'name']).mean().sort_values('marks')
```

```python
# get records that are grouped by by all the columns 

# select term, name, subject, sum(marks) from df 
# group by term, name, subject 


df.groupby(['term','name', 'subject']).sum().sort_values(['term', 'name', 'subject']).head(10)
```
