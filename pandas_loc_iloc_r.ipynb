```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
#sns.get_dataset_names()
df
```

**Specific records/rows**

```python
df.loc[[3]]  # work around to get a tablu
```

```python
df.iloc[[3]]
```

```python
df.loc[[3,4]]
```

```python
df.iloc[[3,4]]
```

**Slicing** 

loc  = [ from_number : to_number ]   - both inclusive <br>
iloc = [ from_number : to_number ]   - from_number inclusive, to_number exclusive


```python
df
```

```python
df.loc[1:4]
```

```python
df.iloc[1:4+1]     # to_number is exclusive, make it 5 to include 4
```

```python
df.loc[2: :3]    # with step 2, alternative rows     ## 1, 2, 3, 4
```

```python
df.iloc[2::3]    # 1, 2, 3
```

```python
df.head(3)
```

```python
df.loc[[0,2], [3,4]]
```

```python
df.iloc[[0,2], [3,4]]
```

```python
df.loc[2:10, :]                         # [      from_row     :  to_row   , from_col    :  to_col       ]
```

```python
df.iloc[2:10+1, :]
```

```python
df.loc[ : , 2:4 ]
```

```python
df.iloc[: ,2:4+1]
```
