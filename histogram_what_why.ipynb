```python
import pandas as pd
import numpy as np
from collections import Counter as c
import matplotlib.pyplot as plt
```

```python
def getBarGraph(x) :
    d = c(x)
    xmin = min(d.keys())
    xmax = max(d.keys())
    plt.figure(figsize=(25,5))
    plt.xlim(xmin-1, xmax+1)
    plt.xticks(range(xmin, xmax+1, 1))
    plt.bar(d.keys(), d.values())
```

```python
x = np.random.randint(1,11, 10)
getBarGraph(x)
```
