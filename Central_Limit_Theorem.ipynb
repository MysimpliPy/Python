<font color='blue'>
    
Central Limit Theorem, a statistical theory, states that as sample sizes get larger, the mean of all samples will<br>
be approximately equal to the mean of the population, and the distribution of means will approach normality, <br>
regardless of whether the population has a uniform, normal,  binomial, or any other distribution.                 <br>
  
    
</font>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n  = 100000
rv = np.random.randint(1,7, n)  # population 
# the integers from 1-6 are getting generated randomly, and the probability of each integer
# getting selected is 1/6 or .16667 or 16.667
d = {}
for i in rv :
    d[i] = d.get(i, 0) + 1
d_sorted = { i : d[i] for i in sorted(d.keys())}
plt.bar(d_sorted.keys(), d_sorted.values())
plt.show()
sample_means = [np.mean(np.random.choice(rv, 30)) for i in range(10000)]
y = plt.hist(sample_means, 10)
z = np.linspace(y[1].min(), y[1].max(), 10)
plt.plot(z, y[0])
plt.show()
```
