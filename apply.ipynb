Pandas apply method : <br>
It passes a function and applies it on every single value of the Pandas series
</font>

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
revised_bill_1 = df['total_bill'].apply(lambda x: x*1.1)
```

```python
df['revised_bill_2'] = df['total_bill'] *1.1
df['revised_bill_3'] = df.total_bill *1.1 
df['revised_bill_4'] = df.loc[:, 'total_bill'] * 1.1
df['revised_bill_5'] = list(map(lambda x: x*1.1, df.total_bill))
```

```python
def addTax(x) :
    return x*1.1
df['revised_bill_6'] = list(map(addTax, df.total_bill))
```

```python
df.apply(lambda row : row[2].lower(), axis=1)
```

```python
df.apply(lambda row : 'male' if row[2] == 'female' else 'female', axis=1)
```

```python
def dfUpper(x) :
    y = []
    for i in range(len(x)) :
        if type(x[i]) == str :
            y.append(x[i].upper())
        else :
            y.append(x[i])
    return y
```

```python
df1 = df.apply(dfUpper, axis=1)
```

```python
df_upper = pd.DataFrame([i for i in df1])
df_upper
```

```python
df_upper.columns = df.columns
```

```python
df_upper
```

```python
def addNumericFields(x) :
    s = 0
    for i in range(len(x)) :
        if isinstance(x[i], int) or isinstance(x[i], float) :
            s = s+x[i]   
    return s
```

```python
df.apply(addNumericFields, axis=1)
```

```python
def mergeFields(x) : 
    s = ''
    for i in range(len(x)) :
        s = s + ',' + str(x[i])
    return s[1:]
```

```python
df.apply(mergeFields, axis=1)
```

```python
def countChar(x): 
    length= 0
    for i in range(len(x)) :
        length = length + len(str(x[i]))
    return length
```

```python
df.apply(countChar, axis=1)
```
