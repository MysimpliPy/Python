```python
import numpy as np
import pandas as pd
```

```python
num = [i for i in range(11,21)]
```

```python
num = np.array(num)
```

```python
square = num*num
cube = num*num*num
log = np.log(num)
sqrt = np.sqrt(num)
```

```python
num
```

```python
square
```

```python
cube
```

```python
log
```

```python
sqrt
```

```python
df = pd.DataFrame(list(zip(num, square, cube, log, sqrt)))
```

```python
df.columns = ['num', 'square', 'cube', 'log', 'sqrt']
```

```python
rows = ['row'+str(i) for i in range(10)]
```

```python
df.index = rows
```

```python
df
```

```python
df
```

```python
df
```

```python
df.loc[['row0', 'row1']]
```

```python
df.iloc[[0, 1]]
```

```python
df.loc['row0': 'row4']    # from a particular row to a paricular 
```

```python
df.iloc[0 : 5]
```
