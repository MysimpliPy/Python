<font size='4' color='blue'>
According to the law of large numbers, if we flip a coin a large number<br>
of times, the average result will be closer to the expected value of 0.5.<br>
</font>


```python
import numpy as np
import pandas as pd
from collections import Counter as C
import matplotlib.pyplot as plt
```

```python
x = np.random.choice(['Head','Tail'], 1000000 )
d = C(x)
d
```

```python
plt.xticks(None)
plt.bar(d.keys(), d.values())
plt.show()
```

```python
np.mean([.5, .5])   # expected value for flipping a coin 
```

```python
jump = [10,100,1000, 10000, 100000]
for i in jump : 
    x = np.random.choice(['Head','Tail'], i) 
    d = C(x)
    plt.xlim(.5, 0,1, 1.5)
    plt.xticks(None)
    plt.bar(d.keys(), d.values())
    plt.show()
    
```

```python
df = pd.DataFrame(x,columns=['HT'])
df['HeadCount'] = np.where(df.HT == 'Head', 1, 0 )
df['TailCount'] = np.where(df.HT == 'Tail', 1, 0 )
df['HeadRatio'] = np.cumsum(df.HeadCount)/(df.index+1)
df['TailRatio'] = np.cumsum(df.TailCount)/(df.index+1)
```

```python
df.tail()
```

```python
plt.figure(figsize=(15, 3))
plt.plot(df.index+1, df.HeadRatio)
plt.plot(df.index+1, df.TailRatio)
plt.show()
```

```python
for i in range(100, len(df), 10000) :
    plt.figure(figsize=(15, 3))
    plt.plot((df.index+1)[:i], df.HeadRatio[:i])
    plt.plot((df.index+1)[:i], df.TailRatio[:i])
    plt.show()
```
