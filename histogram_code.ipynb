```python
import pandas as pd
import numpy as np
from collections import Counter as c
import matplotlib.pyplot as plt
```

```python
def getBarGraph(x) :
    d = c(x)
    xmin = min(d.keys())
    xmax = max(d.keys())
    plt.figure(figsize=(25,5))
    plt.xlim(xmin-1, xmax+1)
    plt.xticks(range(xmin, xmax+1, 1))
    plt.bar(d.keys(), d.values())
```

```python
np.random.seed(1)
x = np.random.randint(1,11, 10)
getBarGraph(x)
```

```python
# this code counts all the datapoints within each and every range of values/bins/buckets 

def getFreq(a, buckets) :
    a = np.sort(a)
    start = 0
    buckets1 = buckets.copy()
    buckets1[-1] = np.PINF
    freq = []
    for b in range(len(buckets1)-1) :
        ctr = 0
        for i in range(start, len(a)) :
            if a[i] >= buckets1[b] and a[i] < buckets1[b+1] :
                ctr += 1
            else :
                if a[i] >= buckets1[b+1] :
                    start = i
                    break
        freq.append(ctr)
    return freq
```
