```python
np.random.seed(1)
x = np.random.randint(1,11, 25)
y = np.random.randint(1,11, 25)
```

```python
df = pd.DataFrame(list(zip(x, y)), columns=['x', 'y'])

#df.head()

centroid = df.loc[[0]]

#centroid

plt.scatter(df.x, df.y)
#plt.scatter(centroid.x, centroid.y, s=200, marker='*')
plt.show()
plt.scatter(df.x, df.y)
plt.scatter(centroid.x, centroid.y, s=200, marker='*')
plt.show()

df['distance'] = np.sqrt((df.x - centroid.loc[0].x)**2  + (df.y - centroid.loc[0].y)**2)

#df.head()

#np.mean(df.x)

#np.mean(df.y)

avg = df[['x', 'y']].mean()

df_avg = pd.DataFrame([avg])

#df_avg

centroid_old = centroid

#id(centroid_old), id(centroid)

centroid = df_avg

#id(centroid), id(df_avg)

plt.scatter(df.x, df.y)
plt.scatter(centroid.x, centroid.y, s=200, marker='*')
plt.show()
```

```python
#####  Machine Learning
#####  K-Means #1 

# Demo on scatter plot and its associated parameters

mn = 1
mx = 10
size = 5
np.random.seed(0)
x = np.random.choice(range(mn, mx+1), size, replace=False)
y = np.random.choice(range(mn, mx+1), size, replace=False)
######################################################################

plt.scatter(x, y)
plt.show()
######################################################################

plt.scatter(x, y)
plt.xlabel('x-axis', fontdict={'c':'blue', 'size':15})
plt.ylabel('y-axis', fontdict={'c':'blue', 'size':15})
plt.show()
######################################################################

plt.scatter(x, y)
plt.xlabel('x axis', fontdict={'c':'blue', 'size':15})
plt.ylabel('y axis', fontdict={'c':'blue', 'size':15})
plt.xlim(0,max(x)+1)
plt.ylim(0,max(y)+1)
plt.show()
######################################################################

plt.scatter(x, y)
plt.xlabel('x axis', fontdict={'c':'blue', 'size':15})
plt.ylabel('y axis', fontdict={'c':'blue', 'size':15})
plt.xlim(0,max(x)+1)
plt.ylim(0,max(y)+1)
plt.xticks(range(0,max(x)+1,1))
plt.yticks(range(0,max(y)+1,1))
plt.show()
######################################################################

plt.scatter(x, y)
plt.xlabel('x axis', fontdict={'c':'blue', 'size':15})
plt.ylabel('y axis', fontdict={'c':'blue', 'size':15})
plt.xlim(0,max(x)+1)
plt.ylim(0,max(y)+1)
plt.xticks(range(0,max(x)+1,1))
plt.yticks(range(0,max(y)+1,1))
plt.hlines(y, 0, x, linestyles='dotted')
plt.vlines(x, 0, y, linestyles='dotted')
plt.show()
######################################################################

plt.figure(figsize=(10,5))
plt.scatter(x, y)
plt.xlabel('x axis', fontdict={'c':'blue', 'size':15})
plt.ylabel('y axis', fontdict={'c':'blue', 'size':15})
plt.xlim(0,max(x)+1)
plt.ylim(0,max(y)+1)
plt.xticks(range(0,max(x)+1,1))
plt.yticks(range(0,max(y)+1,1))
plt.hlines(y, 0, x, linestyles='dotted')
plt.vlines(x, 0, y, linestyles='dotted')
plt.show()
######################################################################
```
